```python
# Ôn lại công thức trước khi code
# Bài toán:
#   Minimize:   w^T × Σ × w       (portfolio variance)
#   Subject to: sum(w) = 1         (tổng tỷ trọng = 100%)
#               w_i >= 0           (long-only, không short)

# Ý nghĩa trực quan với 2 mã đơn giản:
import numpy as np

# Giả sử 2 mã A và B
vol_A = 0.25   # volatility 25%
vol_B = 0.20   # volatility 20%
corr  = 0.30   # tương quan 30%

# Covariance Matrix thủ công
cov = np.array([
    [vol_A**2,         corr*vol_A*vol_B],
    [corr*vol_A*vol_B, vol_B**2        ]
])
print("Covariance Matrix:")
print(cov.round(4))

# Test nhiều tỷ trọng, tìm điểm có variance thấp nhất
variances = {}
for w1 in [i/10 for i in range(0, 11)]:
    w2 = 1 - w1
    w  = np.array([w1, w2])
    var = float(w @ cov @ w)
    variances[w1] = var

best_w = min(variances, key=variances.get)
print(f"\nDuyệt thủ công — Min variance tại w_A = {best_w:.1f}, w_B = {1-best_w:.1f}")
print(f"Min variance = {variances[best_w]:.4f} → Volatility = {variances[best_w]**0.5:.2%}")
print(f"\nNếu chỉ dùng B (vol thấp hơn): {vol_B:.2%}")
print(f"→ Danh mục mix còn thấp hơn cả mã ít rủi ro nhất! Đây là lợi ích diversification.")
```

```python
from scipy.optimize import minimize

# Bài toán mẫu: minimize x^2 + y^2 với constraint x + y = 1
def objective(vars):
    x, y = vars
    return x**2 + y**2

constraints = [
    {'type': 'eq', 'fun': lambda v: v[0] + v[1] - 1}  # x + y = 1
]
bounds = [(-10, 10), (-10, 10)]   # không cần restrict ở đây
x0    = [0.5, 0.5]                 # initial guess

result = minimize(objective, x0, method='SLSQP',
                  bounds=bounds, constraints=constraints)

print("=== Ví dụ minimize x^2 + y^2 với x + y = 1 ===")
print(f"x = {result.x[0]:.4f}, y = {result.x[1]:.4f}")
print(f"Tổng: {sum(result.x):.4f} (phải = 1)")
print(f"Objective: {result.fun:.4f}")
print(f"Thành công: {result.success} | Message: {result.message}")

# Giải thích: minimum của x^2+y^2 với x+y=1 là tại x=y=0.5
# → portfolio tương tự: minimize variance với sum=1 cho optimal weights
```

```python
# scipy trả về OptimizeResult object — cần hiểu các field
print("Các field quan trọng của result:")
print(f"  result.x       : {result.x}           ← optimal solution (weights)")
print(f"  result.fun     : {result.fun:.6f}      ← objective value tại optimal")
print(f"  result.success : {result.success}           ← True nếu converged")
print(f"  result.message : {result.message}")
print(f"  result.nit     : {result.nit}              ← số iterations")
```

```python
test_cases = {
    '5 mã đa ngành'  : ['VCB', 'VNM', 'HPG', 'FPT', 'MWG'],
    '5 mã ngân hàng' : ['VCB', 'BID', 'CTG', 'MBB', 'ACB'],
    '5 mã low-vol'   : ['VCB', 'KDC', 'VNM', 'SAB', 'REE'],
    '10 mã VN30'     : ['VCB','BID','CTG','TCB','MBB',
                         'FPT','VNM','HPG','GAS','KDC'],
    '30 mã VN30'     : None,  # full VN30
}

from data_loader import VN30_TICKERS

rows = []
for label, tickers in test_cases.items():
    t = tickers if tickers else VN30_TICKERS
    try:
        r = min_variance_portfolio(t, START, END)
        rows.append({
            'Tổ hợp'    : label,
            'N mã'      : len(t),
            'MVP Vol'   : f"{r['port_volatility']:.2%}",
            'MVP Return': f"{r['port_return']:.2%}",
            'MVP Sharpe': f"{r['sharpe_ratio']:.3f}",
            'EW Vol'    : f"{r['equal_weights_vol']:.2%}",
            'Giảm Vol'  : f"{r['improvement_pct']:.1f}%",
            'N active'  : int((r['weights'] > 0.001).sum()),
            'Converged' : '✅' if r['success'] else '❌',
        })
    except Exception as e:
        rows.append({'Tổ hợp': label, 'N mã': '?', 'MVP Vol': f'ERROR: {e}',
                     'MVP Return':'', 'MVP Sharpe':'', 'EW Vol':'',
                     'Giảm Vol':'', 'N active':'', 'Converged':'❌'})

import pandas as pd
comparison = pd.DataFrame(rows)
print("\n=== Bảng so sánh tổ hợp danh mục ===")
print(comparison.to_string(index=False))
```
